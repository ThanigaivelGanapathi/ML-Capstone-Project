```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
```

```python
dataset = pd.read_csv("Loan_default.csv")
```

```python
dataset
```

```python
new_data = dataset.copy()
```

```python
new_data
```

```python
new_data.drop(["LoanID"],axis = 1,inplace = True)
```

```python
new_data
```

```python
new_data.describe()
```

```python
new_data["Default"].value_counts()
```

```python
yes_data = new_data[new_data["Default"] == 1]
no_data  = new_data[new_data["Default"] == 0]
```

```python
yes_data
```

```python
min_size = min(len(yes_data), len(no_data))
min_size
```

```python
min_yes_data = yes_data.sample(min_size,random_state = 42)
min_no_data  = no_data.sample(min_size,random_state = 42)
```

```python
min_no_data
```

```python
balanced_data = pd.concat([min_yes_data,min_no_data])
```

```python
balanced_data
```

```python
balanced_data=  shuffle(balanced_data, random_state=42)
```

```python
balanced_data
```

```python
reduced_data = balanced_data.sample(n=1000,random_state = 42)
```

```python
reduced_data
```

```python
reduced_data["Default"].value_counts()
```

```python
balanced_data["Default"].value_counts()
```

```python
final_yes = balanced_data[balanced_data['Default'] == 1].sample(n=500, random_state=42)
final_no = balanced_data[balanced_data['Default'] == 0].sample(n=500, random_state=42)
```

```python
new_reduced_data = pd.concat([final_yes, final_no])
```

```python
new_reduced_data = shuffle(new_reduced_data, random_state = 42)
```

```python
new_reduced_data["Default"].value_counts()
```

```python
new_reduced_data
```

```python
new_reduced_data.to_csv("Balanced_Dataset.csv",index = False)
```
